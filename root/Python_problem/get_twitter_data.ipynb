## Function to get data

```python
def get_twitter_data(username):
    
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret) 

    #set access 
    auth.set_access_token(access_key, access_secret) 
    #create tweepy api 
    api = tweepy.API(auth) 

     
    number_of_tweets=1000
    
    #get tweets
    tweets = api.user_timeline(username, tweet_mode='extended',count=100)
    
    #get data as string
    jsondatastr = [json.dumps(tweet._json) for tweet in tweets]
    
    #load string data as json
    jsondata = [json.loads(json_str) for json_str in jsondatastr]

    #get tweet texts
    tweettext = [tweet['full_text'] for tweet in jsondata] # CSV file created  
    
    #get tweet dates
    tweetdate = [time.strftime('%Y-%m-%d', time.strptime(tweet['created_at'],'%a %b %d %H:%M:%S +0000 %Y')) for tweet in jsondata]
    
    #get tweet time
    tweettime = [time.strftime('%H:%M:%S', time.strptime(tweet['created_at'],'%a %b %d %H:%M:%S +0000 %Y')) for tweet in jsondata]

    #get tweet retweet count
    retweetcount = [tweet['retweet_count'] for tweet in jsondata]

    #get tweet like count
    likecount = [tweet['favorite_count'] for tweet in jsondata]

    #check if tweet contains photo
    photo = []
    for tweet in jsondata:
        if ('media' in tweet['entities']):
            if (tweet['entities']['media'][0]['type']=='photo'):
                photo.append(1)
            else:
                photo.append(0)
        else:
            photo.append(0)

    #create dataframe        
    df = pd.DataFrame()
    df['date'] = tweetdate
    df['time'] = tweettime
    df['tweet'] = tweettext
    df['likes'] = likecount
    df['retweets'] = retweetcount
    df['photo'] = photo

    return df
```

```python
#create twitter dataframe for midasIIITD or any user
df = get_twitter_data('midasIIITD')
```
